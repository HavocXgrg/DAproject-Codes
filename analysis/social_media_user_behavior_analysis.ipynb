##### Data Type Optimization

```python
df['account_join_date'] = pd.to_datetime(df['account_join_date'], errors='coerce')
# Convert categorical columns
categorical_cols = [
    'gender', 'country', 'profession', 'primary_platform',
    'preferred_device', 'preferred_content_type'
]
df[categorical_cols] = df[categorical_cols].astype('category')
# Convert numeric columns that should be integers
int_cols = [
    'platforms_used_count', 'followers_count', 'following_count',
    'posts_per_week', 'likes_given_per_day', 'comments_per_day',
    'shares_per_day', 'dms_sent_per_day'
]
df[int_cols] = df[int_cols].astype('int')
```

```python
plt.figure(figsize=(8,6))
sns.countplot(data=df, x='gender', hue='gender')
plt.title("Distribution of gender of people",  fontsize=14, fontweight='bold')
```

The countplot shows that the dataset is predominantly composed of Female and Male individuals, with very few entries for Non-binary and Prefer not to say categories, indicating a gender imbalance.

```python
plt.figure(figsize=(12, 5))
# Histogram on the left
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='age', bins=20, kde=True)
plt.title("Age Distribution (Histogram)", fontsize=14, fontweight='bold')
plt.xlabel("Age")
plt.ylabel("Count")

# Boxplot on the right
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='age',)
plt.xticks([20, 25, 30, 35, 40, 45, 50])
plt.title("Age Distribution (Boxplot)", fontsize=14, fontweight='bold')
plt.xlabel("Age")

plt.tight_layout()
plt.show()
```

The histplot is right‑skewed, meaning there are fewer older individuals beyond 40.
The boxplot shows the median age is around 27, and a few outliers appear above 45, indicating some older participants.
Together, these plots reveal that the dataset is dominated by younger individuals, with limited representation of older ages.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='gender', y='age', data=df, hue='gender')
plt.title("Age Distribution for each gender", fontsize=14, fontweight='bold')
plt.yticks([20, 25, 30, 35, 40, 45, 50])
plt.show()
```

This chart shows the age distribution across gender groups.
It reveals that median ages are similar for all genders, with males showing slightly more variability and a few older outliers, while other groups have a more compact age range.

### Feature Engineering

##### Categories of Age feature

```python
# creatign a new column age category

bins = [0, 12, 19, 35, 60, 100]
labels = ['Child (0-12)', 'Teen (13-19)', 'Young Adult (20-35)', 'Adult (36-60)', 'Senior (60+)']

# Create new column
df['age_category'] = pd.cut(df['age'], bins=bins, labels=labels)
```

```python
plt.figure(figsize= (12,6))
sns.countplot(data=df, x= 'age_category', palette='Set2')
plt.title("Distribution of Age categories", fontsize=14, fontweight='bold')
plt.xlabel("Age Categories", fontsize=12, fontweight='bold')
plt.ylabel("Count", fontsize=12, fontweight='bold')
plt.show()
```

##### Total Interaction per day Feature

```python
df['total_daily_interactions'] = (
    df['likes_given_per_day'] +
    df['comments_per_day'] +
    df['shares_per_day'] +
    df['dms_sent_per_day']
)
```

```python
sns.histplot(data=df, x='total_daily_interactions', bins=30, kde=True )
```

The shape is roughly bell‑like, centered around 30–40 interactions per day. This means most users are moderately active, with fewer very low (< 20) or very high (> 50) interaction counts.

#### Spend-to-Click Ratio and Ad Engagement

```python

df['spend_to_click_ratio'] = df['monthly_spend_via_social_usd'] / (df['ad_click_rate'] + 0.001)

# Quick check
print(df[['monthly_spend_via_social_usd', 'ad_click_rate', 'spend_to_click_ratio']].head())
    
```

```python

sns.regplot(x='spend_to_click_ratio', y='monthly_spend_via_social_usd', data=df)
```


The regression plot shows a **positive relationship** between monthly social media spending and the spend-to-click ratio, indicating that higher spenders tend to have higher ratios, meaning they spend more per unit of ad engagement. 
This helps identify **ad efficiency patterns**, distinguishing high-value spenders from frequent clickers.


##### Extracting year, month and day form the date

```python
# Extract Year
df['join_year'] = df['account_join_date'].dt.year

# Extract Month
df['join_month'] = df['account_join_date'].dt.month

# Extract Day 
df['join_day'] = df['account_join_date'].dt.day

# Extract Day Name 
df['join_day_name'] = df['account_join_date'].dt.day_name()

# Quick inspection of featured columns
print(df[['account_join_date', 'join_year', 'join_month', 'join_day', 'join_day_name']].head())
```

```python
plt.figure(figsize=(10, 6))

# 1. Automatically calculate the average hours per profession and sort them from highest to lowest
profession_order = df.groupby('profession')['daily_usage_hours'].mean().sort_values(ascending=False).index

# 2. Plot a simple horizontal bar chart
sns.barplot(data=df, x='daily_usage_hours', y='profession', order=profession_order, errorbar=None)

# 3. Add simple labels
plt.title('Which Profession Spends the Most Time on Social Media?', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Average Daily Usage (Hours)', fontsize=12)
plt.ylabel('Profession', fontsize=12)

plt.tight_layout()
plt.show()
```
