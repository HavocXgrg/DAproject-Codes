**Handling Null Values, converting datatypes**

```python
# Fill text-based columns
df["motivation"] = df["motivation"].fillna("No official citation")
df["birth_city"] = df["birth_city"].fillna("Unknown")
df["birth_country"] = df["birth_country"].fillna("Unknown")
df["sex"] = df["sex"].fillna("Not Specified")
df["organization_name"] = df["organization_name"].fillna("Independent")
df["organization_city"] = df["organization_city"].fillna("Unknown")
df["organization_country"] = df["organization_country"].fillna("Unknown")

df["death_city"] = df["death_city"].fillna("Unknown")
df["death_country"] = df["death_country"].fillna("Unknown")
# Handle laureate_id separately
df["laureate_id"] = df["laureate_id"].fillna(-1).astype("Int64")
```

```python
# Converting date columns to datetime format
# errors='coerce" ensures that invalid or missing values become NaT (Not a Time)
df['birth_date'] = pd.to_datetime(df['birth_date'], errors='coerce')
df["death_date"] = pd.to_datetime(df["death_date"], errors="coerce")
```

```python
# converting fractions which was object to numeric float value
df['prize_share'] = df['prize_share'].map(
    lambda x: float(x.split('/')[0]) / float(x.split('/')[1]) if isinstance(x, str) else x
)
```

```python
# converting the types of columns which was object into categorical
categorical_cols = [
    "category", "laureate_type", "sex",
    "birth_country", "organization_country",
    "organization_name", "organization_city",
    "death_country", "death_city"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")
```

# Feature Engineering

```python
# Extracting only year for numerical math.
df['birth_year'] = df['birth_date'].dt.year.astype('Int64')
df['death_year'] = df['death_date'].dt.year.astype('Int64')
```

```python
# Histplot to show the distribution of birth and death year
fig, axes = plt.subplots(1, 2, figsize=(12,6))

sns.histplot(df['birth_year'], bins=10, kde=True, ax=axes[0], color='blue')
axes[0].set_title("Birth Year")

sns.histplot(df['death_year'], bins=10, kde=True, ax=axes[1], color='red')
axes[1].set_title("Death Year")

plt.tight_layout()
plt.show()
```

```python
# FEATURE 1: Age of the Laureate at the time of receiving the prize
df['award_age'] = df['year'] - df['birth_year']
```

```python
#FEATURE 2: Lifespan / Longevity of individual laureates
df['lifespan'] = df['death_year'] - df['birth_year']
```

```python
#FEATURE 3: Boolean flag identifying if the prize was shared or solo
df['is_shared'] = df['prize_share'] < 1.0
```

```python
# Verifying engineered feature
df[['full_name', 'award_age', 'lifespan', 'is_shared']].head()
```

```python
#Box-plot to know spread of ages at which awards were received
plt.figure(figsize=(8, 4))
sns.boxplot(x=df['award_age'])
plt.title("Distribution of Award Age", fontsize=14, fontweight='bold')
plt.xlabel("award_age", fontweight='bold')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="category", y="award_age", data=df, hue="category", palette="Set2", legend=False)
plt.title("Award Age by Category")
plt.xticks(rotation=45)
plt.show()
```
